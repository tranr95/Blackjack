```python
class Deck:
    
    def __init__(self):
        self.all_cards = []
        for suit in suits:
            for rank in ranks:
                self.all_cards.append(Card(suit,rank))
    
    def shuffle(self):
        random.shuffle(self.all_cards)
        
    def deal(self):
        return self.all_cards.pop()
    
    def add_card(self,card):
        self.cards.append(card)
    
  
        
```

```python
class Hand:
    
    def __init__(self,user):
        self.user = user
        self.cards = []
        self.value = 0
        self.aces = 0
        
    def __str__(self):
        print('{}'.format(self.user))
        
    
    def add_card(self,card):
        if card.values == 11:
            self.aces = self.aces + 1
            self.cards.append(card)
            self.value = self.value + card.values
        
        else:
            self.cards.append(card)
            self.value = self.value + card.values
            
    #you will adjust aces after every hit. previous aces will have their value already adjusted prior to current ace
    def adjust_for_ace(self):
        if self.aces >= 1 and self.value > 21:
            self.value = self.value - 10
            self.aces = self.aces - 1
            return self.value
        
        else:
            pass
    
    
```

```python
class Chips:
    
    def __init__(self,player,bankroll):
        self.player = player
        self.bankroll = bankroll
        
        #BET MUST BE LOWER THAN BANKROLL. LOOP IT BACK TO THE INPUT IF BET IS GREATER THAN BANKROLL
    def take_bet(self,bet):
        self.bankroll = self.bankroll - bet
        return bet
    
    def win_bet(self,bet):
        self.bankroll = self.bankroll + bet + bet
        return self.bankroll
   
    def lose_bet(self,bet):
        self.bankroll = self.bankroll
        return self.bankroll
    
    def push_bet(self,bet):
        self.bankroll = self.bankroll + bet
        return self.bankroll
```

```python
gamedeck = Deck()
player = Hand('Player')
dealer = Hand('Dealer')
bukka = Chips('Bukka',10000)
```

```python
def place_bet(bank):
    #BET MUST BE LOWER THAN BANKROLL. LOOP IT BACK TO THE INPUT IF BET IS GREATER THAN BANKROLL
    placebet = True
    while placebet:
        bet = int(input('Please place your bets: '))
        if bet > bank.bankroll:
            print('Insufficient Funds')
            continue
        
        elif bank.bankroll < 0:
            print('BANKROLL = $0')
            placebet = False
        
        else:
            print('Bet confirmed')
            print('Your wager amount: {}'.format(bet))
            return bet
            placebet = False
            
```

```python
def hit(Hand):
    Hand.add_card(gamedeck.deal())
    #find a way to adjust for ace only if last card is ace
    if Hand.aces == 1:
        Hand.adjust_for_ace()
    else:
        pass
    print(Hand.cards[-1])
    print('{} Hand Value: {}'.format(Hand.user,Hand.value))
```

```python
def hit_or_stand(Hand):
    while Hand.value < 21:
        action = input('Hit or Stand: ')
        if action.lower() == 'hit':
            hit(Hand)
            continue
        else:
            print('{} Hand Value: {}'.format(Hand.user,Hand.value))
            return False
```

```python
#END GAME SCENARIOS
#pLayer wins
def win(player,dealer):
    if player > dealer:
        print('Player Wins!')
        return 'win'
#dealer busts
    elif dealer > 21:
        print('Dealer Busts!')
        print('Player Wins!')
        return 'win'
    
#player push
    elif player == dealer:
        print('Player Push!')
        return 'push'
        
 #player busts
    elif player > 21:
        print('Player Busts!')
        print('Dealer Wins!')
        return 'lose'
    
#dealer wins
    else:
        print('Dealer Wins!')
        return 'lose'
```

```python
#GAME
def Blackjack():
    gamerun = True
    while gamerun:
        bet = bukka.take_bet(place_bet(bukka))
        gamestart()
        show_some(player,dealer)
    
    #Player Turn
   
        playerturn = True
        while playerturn:
            hit_or_stand(player)
            
            #player bust
            if player.value > 21:
                player_busts()
                playerturn = False
            #check for blackjack
            elif player.value == 21 and len(player.cards) == 2:
                print('Blackjack!')
                bukka.win_bet(bet)
                playerturn = False  
            
            else:
                playerturn = False
              
    #Dealers Turn
        #dealer shows card and hits until >= 17
        show_hand(dealer)
        print('{} Hand Value: {}'.format(dealer.user,dealer.value))
        dealerturn = True
        while dealerturn:
            if dealer.value < 17 and player.value < 21:
                hit(dealer)
                continue
            else:
                dealerturn = False
        
    #compare two hands to determine winner
        wincon = True
        while wincon and player.value < 21:
            if win(player.value,dealer.value) == 'win':
                bukka.win_bet(bet)
                wincon = False
            elif win(player.value,dealer.value) == 'push':
                bukka.push_bet(bet)
                wincon = False
            else:
                bukka.lose_bet(bet)
                wincon = False
            
        print('{}s Bankroll: {}'.format(player.user,bukka.bankroll))
    
    #run replay
        
        gamerun = False
        replay()
        
```
